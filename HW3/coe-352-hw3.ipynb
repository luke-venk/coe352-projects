## Explicit Euler

```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 0.01
T = 10 / alpha  # arbitrary T
#               0    1     2     3     4      5
time_steps = [0.01, 1.0, 10.0, 50.0, 100.0, 200.0]

def forward_euler(deltaT: float) -> np.ndarray:
    # Numerical solution: Forward Euler
    n = int(T / deltaT)
    t = np.linspace(0, T, n + 1)
    y = np.zeros(n + 1)
    y[0] = 1
    for i in range(n):
        y[i + 1] = y[i] * (1 - alpha * deltaT)
        
    # Analytical solution
    t_exact = np.linspace(0, T, 500)
    
    # Plot
    plt.figure()
    plt.plot(t_exact, np.exp(-alpha * t_exact), label='Exact Solution')
    plt.plot(t, y, '--', label='Numerical Solution')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Forward Euler with deltaT = {deltaT}')
    plt.legend()
    plt.show()

def backward_euler(deltaT: float) -> np.ndarray:
    # Numerical solution: Backward Euler
    n = int(T / deltaT)
    t = np.linspace(0, T, n + 1)
    y = np.zeros(n + 1)
    y[0] = 1
    for i in range(n):
        y[i + 1] = y[i] / (1 + alpha * deltaT)
        
    # Analytical solution
    t_exact = np.linspace(0, T, 500)
    
    # Plot
    plt.figure()
    plt.plot(t_exact, np.exp(-alpha * t_exact), label='Exact Solution')
    plt.plot(t, y, '--', label='Numerical Solution')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Backward Euler with deltaT = {deltaT}')
    plt.legend()
    plt.show()
    
def trapezoid_method(deltaT: float) -> np.ndarray:
    # Numerical solution: Trapezoid
    n = int(T / deltaT)
    t = np.linspace(0, T, n + 1)
    y = np.zeros(n + 1)
    y[0] = 1
    for i in range(n):
        y[i + 1] = y[i] * (1 - 0.5 * alpha * deltaT) / (1 + 0.5 * alpha * deltaT)
        
    # Analytical solution
    t_exact = np.linspace(0, T, 500)
    
    # Plot
    plt.figure()
    plt.plot(t_exact, np.exp(-alpha * t_exact), label='Exact Solution')
    plt.plot(t, y, '--', label='Numerical Solution')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Trapezoid Method with deltaT = {deltaT}')
    plt.legend()
    plt.show()
```

```python
deltaT = time_steps[2]

forward_euler(deltaT)
backward_euler(deltaT)
trapezoid_method(deltaT)
```
